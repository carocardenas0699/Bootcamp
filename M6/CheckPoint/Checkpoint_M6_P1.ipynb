```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.datasets import make_moons

# Generar datos de ejemplo
X, y = make_moons(n_samples=200, noise=0.25, random_state=0)

# Definir el espacio de búsqueda de hiperparámetros
param_grid = {
    'kernel': ['sigmoid', 'rbf'],
    'C': [0.01, 0.1, 1, 10, 100, 1000],
    'gamma': [0, 0.1, 1, 10, 100],
    'coef0': [0, 0.1, 1]
}

# Instanciar el modelo SVM
modelo_svm = SVC()

# Instanciar GridSearchCV
grid_search = GridSearchCV(modelo_svm, param_grid, cv=3, scoring='accuracy', n_jobs=-1)

```

```python
from sklearn.model_selection import train_test_split
import pandas as pd

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
grid_search.fit(X_train,y_train)

print("Mejores hiperparámetros: "+str(grid_search.best_params_))
print("Mejor Score: "+str(grid_search.best_score_)+'\n')

scores = pd.DataFrame(grid_search.cv_results_)
scores
```
