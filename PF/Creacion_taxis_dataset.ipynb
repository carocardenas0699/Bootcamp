```python
import pandas as pd
import glob
import os
```

# Creacion de taxis.parquet

Este archivo contiene el proceso llevado a cabo para la creacion del dataset de **taxis** que se empleara para el analisis. Estos datos fueron extraidos de la pagina oficial del NYC TLC

```python
archivos_parquet = glob.glob(os.path.join('parquet/', 'yellow*.parquet'))

yellow_taxis = pd.concat([pd.read_parquet(archivo) for archivo in archivos_parquet])
yellow_taxis
```

```python
yellow_taxis.to_parquet('yellow_taxis.parquet')
```

```python
archivos_parquet = glob.glob(os.path.join('parquet/', 'green*.parquet'))
green_taxis = pd.concat([pd.read_parquet(archivo) for archivo in archivos_parquet])
green_taxis.rename(columns={'lpep_pickup_datetime':'tpep_pickup_datetime', 'lpep_dropoff_datetime':'tpep_dropoff_datetime'},inplace=True)
green_taxis
```

```python
green_taxis.to_parquet('green_taxis.parquet')
```

```python
yellow_taxis['type']='yellow'
green_taxis['type']='green'
```

```python
taxis = pd.concat([yellow_taxis,green_taxis], ignore_index=True)
taxis
```

```python
taxis.to_parquet('taxis.parquet')
```
