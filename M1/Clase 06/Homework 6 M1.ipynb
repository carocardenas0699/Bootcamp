```python
import string as st
```

```python
def hash(val):

    mayu=dict(zip(list(st.ascii_uppercase),list(st.ascii_uppercase)[::-1]))
    minu=dict(zip(list(st.ascii_lowercase),list(st.ascii_lowercase)[::-1]))
    cad = enumerate(repr(val), start=1)
    cod = '' 

    for i in cad:

        if i[1] != "'":

            if str(i[1]).isupper():
                cod+=mayu[i[1]]
            elif str(i[1]).islower():

                cod+=minu[i[1]]
        
    return cod
                
```

```python
hash('PatisimO')
```

```python
h=[]
for i in d_filtrado['Código de país']:
    h.append(hash(i))

d_filtrado['Clave_Hash']=h
d_filtrado
```

```python
def hash_function(key):
    return sum(index * ord(character) for index, character in enumerate(repr(key), start=1))
```

```python
print(hash_function('pato'))
print(hash_function('tapo'))
```
