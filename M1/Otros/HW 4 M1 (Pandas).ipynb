A partir del CSV hospitales2.csv, generar un archivo CSV de salida, que contenga los siguientes campos en este orden:<br>
latitude<br>
longitude<br>
name<br>
label<br>
Con los correspondientes datos extraídos del CSV original, el campo name tiene que corresponder con la dirección del hospital, y el campo label con el nombre del hospital.<br>
Ejemplo de visualización en: https://www.gpsvisualizer.com

```python
import pandas as pd
```

```python
df = pd.read_csv('hospitales2.csv')
df.head(3)
```

```python
def extraer_coordenadas(columna):
    for i in columna:
        i.split()
```

```python
df_salida  = df.WKT.str.slice(7, -1).str.split(expand=True)
```

```python
df_salida.head(2)
```

```python
def extract_coordinates(wkt):
    wkt = wkt.replace("POINT (", "").replace(")", "")
    longitude, latitude = wkt.split()
    return float(latitude), float(longitude)
```

```python
df['latitude'], df['longitude'] = zip(*df['WKT'].apply(extract_coordinates))
```

```python
df.head(2)
```

```python
latitud_longitud = [coordenadas.split() for coordenadas in df['WKT']]
latitude = [float(elemento[1].strip('(')) for elemento in latitud_longitud]
longitude = [float(elemento[2].strip(')')) for elemento in latitud_longitud]
```

```python
lat=[]
lon=[]
for i in df['WKT']:

    d=i.replace('POINT (','').replace(')','').split(' ')

    lat.append(d[1])
    lon.append(d[0])
```

```python
import csv
import os
```

```python
datos = csv.reader(open('hospitales2.csv', encoding = 'utf-8'))

nuevo_archivo = open('hospitalesCSV.csv', 'w', encoding='utf-8', newline='')

escribir = csv.writer(nuevo_archivo)

next(datos)
escribir.writerow(['latitude','longitude','name','label'])
for lines in datos:
    coordenadas = lines[0][7:-1].split()
    escribir.writerow([coordenadas[1], coordenadas[0], lines[7], lines[2]])
    print(lines)

nuevo_archivo.close()
```
