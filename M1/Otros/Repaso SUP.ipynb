## Numpy

Creen un array de numpy con 10 numeros aleatorios entre 0 y 100

```python
a=np.random.randint(0,100,10)
print(a)
```

Filtra y muestra todos los numeros menores de 50 del array anterior

```python
menos_50=a[a<50]
print('Solo menores de 50:',menos_50)
```

```python
mask=((a)<=50)
print('Solo menores de 50:',a[mask])
```

Suma, resta, mult y div en arrays

```python
print('Original:',a)

s=a+5
print('Suma:',s)

r=a-5
print('Resta',r)

m=a*2
print('Multiplicacion',m)

d=a/3
print('Division',d)
```

Usen una mascara para cambiar todos los numeros mayores de 50 a 0

```python
a=np.random.randint(0,100,10)
print('Original:',a)
mask=(a>50)
a[mask]=0
print('Menores de 50 a 0:',a)
```

Crear nueva columna que se llame ganancias multiplicando las columnas plays y price

```python
data["Price (Euros) "] = data["Price (Euros) "].replace("Free to play", "0") #Se debe reemplazar los str a 0
data["Price (Euros) "] = data["Price (Euros) "].str.replace(",", ".") #Cambiar ',' por '.' para poder pasarlo a float

data["Plays"] = data["Plays"].astype(float)
data["Price (Euros) "] = data["Price (Euros) "].astype(float)
```

```python
data['Ganancias']=data['Plays']*data['Price (Euros) '] #Se calcula la columna de ganancias
```
