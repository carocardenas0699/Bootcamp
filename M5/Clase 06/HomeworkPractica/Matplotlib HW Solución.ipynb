# Ejercicios

Sigue las instrucciones para recrear los gráficos utilizando estos datos:

## Los datos

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

**Importa matplotlib.pyplot as plt y establece %matplotlib inline si estás usando el cuaderno jupyter. ¿Qué comando usas si no estás usando el cuaderno jupyter?**

```python
import matplotlib.pyplot as plt
%matplotlib inline
# plt.show() for non-notebook users
```

## Ejercicio 1

** Sigue estos pasos: **
* ** Crea un objeto de figura llamado fig usando plt.figure() **
* ** Usa add_axes para agregar un eje al lienzo de la figura en [0,0,1,1]. Llama a este nuevo eje ax. **
* ** Grafica (x,y) en esos ejes y establece las etiquetas y títulos para que coincidan con el gráfico a continuación: **

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

## Ejercicio 2
** Crea un objeto de figura y pon dos ejes en él, ax1 y ax2. Ubicado en [0,0,1,1] y [0.2,0.5,.2,.2] respectivamente.**

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

** Ahora grafíca (x,y) en ambos ejes. Y llama a tu objeto de figura para mostrarlo. **

```python
ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')


ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')

fig # Show figure object
```

## Ejercicio 3

** Crea el siguiente gráfico añadiendo dos ejes a un objeto de figura en [0,0,1,1] y [0.2,0.5,.4,.4] respectivamente. **

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

** Ahora usa x, y, y z para recrear el gráfico a continuación. Observa los límites de x e y en el gráfico insertado: **

```python
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')


ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

fig
```

## Ejercicio 4

** Usa plt.subplots(nrows=1, ncols=2) para crear el gráfico a continuación. **

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)
```

** Ahora plotea (x,y) y (x,z) en los ejes. Juega con el ancho de línea y el estilo **

```python
axes[0].plot(x,y,color="blue", lw=3, ls='--')
axes[1].plot(x,z,color="red", lw=3, ls='-')
fig
```

** Mira si puedes redimensionar el gráfico añadiendo el argumento figsize() en plt.subplots() copiando y pegando tu código anterior. **

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,color="blue", lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```

** Ahora Usa solo los ejes , ax1 y ax2. **

```python
fig,(ax1,ax2) = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

ax1.plot(x,y,color="blue", lw=5)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x,z,color="red", lw=3, ls='--')
ax2.set_xlabel('x')
ax2.set_ylabel('z')

plt.tight_layout()

plt.show()
```
