## ¿Cuál es la importancia de la visualización de datos?

La visualización de datos es crucial debido a su capacidad para proporcionar una comprensión profunda de tendencias, valores y patrones dentro de los datos. A través de representaciones gráficas, la información se vuelve más accesible y procesable.

### Permite mayor contexto

Imagina que posees datos y tienes la habilidad de realizar análisis estadísticos, como cálculos de desviación estándar o correlación. Sin embargo, esto puede no ofrecer una visión completa de los datos. Por ejemplo, al visualizarlos gráficamente, utilizando un conjunto de datos como Datasaurus, se pueden identificar representaciones diversas a pesar de compartir la misma información.

### Permite hallazgos en nuestros datos

La visualización de datos facilita la detección de hallazgos de manera directa y sencilla. Algunos ejemplos incluyen la capacidad para:

- Identificar tendencias.
- Descubrir comportamientos.
- Detectar valores atípicos (outliers).

### Permite mayor claridad en el código

Además, la visualización de datos se utiliza en campos como la ciencia de datos orientada a machine learning e inteligencia artificial, como en las curvas de aprendizaje. Esto es fundamental en el entrenamiento de algoritmos, ya que las visualizaciones simplifican el entendimiento de aspectos complejos.

## Librerías para visualización de datos

En el ámbito de la visualización de datos, dos librerías sobresalen como fundamentales:

- **Matplotlib**
  - Emula comandos de Matlab.
  - Hace uso de Numpy.
  - Escrita en Python.
  - Facilidad y simplicidad de uso.
  - Personalización y rapidez en su implementación.

- **Seaborn**
  - Construida sobre Matplotlib.
  - Ofrece mayor abstracción y comodidad en la creación de gráficos.

### plt.subplot(nrows, ncols, plot_number)

un subplot es un gráfico dentro de otro gráfico 


| Parámetro   | Descripción                                                            |
|-------------|------------------------------------------------------------------------|
| `nrows`     | Número de filas de subplots en el grid.                                 |
| `ncols`     | Número de columnas de subplots en el grid.                              |
| `index`     | Índice del subplot actual (empieza desde 1 y avanza en filas).         |
| `**kwargs`  | Argumentos adicionales que se pasan a la función `add_subplot()`.       |
| `projection` | Proyección de los subplots (por ejemplo, `'3d'` para gráficos 3D).     |

```python
# Crea un subplot con 1 fila, 2 columnas y selecciona el primer subplot.
plt.subplot(1, 2, 1)

# Plotea una línea roja discontinua en el primer subplot.
plt.plot(x, y, 'r--')

# Crea un subplot con 1 fila, 2 columnas y selecciona el segundo subplot.
plt.subplot(1, 2, 2)

# Genera un histograma de los datos y lo muestra en el segundo subplot.
plt.hist(y)

# Muestra los subplots creados.
plt.show()
```

```python
# Crea un grid de subplots con 1 fila y 2 columnas, y selecciona el primer subplot.
plt.subplot(1, 2, 1)

# Crea un gráfico de línea con datos x e y, usando línea roja discontinua.
plt.plot(x, y, 'r--')

# Crea otro gráfico de línea con datos y y x, usando línea azul punteada.
plt.plot(y, x, 'b:')


# Cambia al segundo subplot en el grid (1 fila y 2 columnas).

plt.subplot(1, 2, 2)

# Crea un gráfico de pastel (pie chart) basado en los datos y.
plt.pie(y)

# Muestra los subplots en la visualización.
plt.show()
```

## plt.figure()

Figura es el contenedor de todos los elementos de la gráfica 

| Parámetro          | Descripción                                                                          |
|--------------------|--------------------------------------------------------------------------------------|
| `num`              | Número de la figura. Puede ser usado para hacer referencia a la figura más tarde. |
| `figsize`          | Tamaño de la figura en pulgadas (ancho, alto) como una tupla.                      |
| `dpi`              | Resolución en puntos por pulgada (dots per inch).                                  |
| `facecolor`        | Color de fondo de la figura.                                                        |
| `edgecolor`        | Color del borde de la figura.                                                       |
| `frameon`          | Si la figura debe tener un borde (True) o no (False).                               |
| `clear`            | Si se debe borrar el contenido de la figura antes de crearla (True) o no (False).  |
| `tight_layout`     | Ajustar automáticamente los márgenes y espacios en blanco para que quepa el contenido.|


| Comando                | Descripción                                                                                                 |
|------------------------|-------------------------------------------------------------------------------------------------------------|
| `fig = plt.figure()`   | Crea una nueva figura vacía.                                                                               |
| `fig.add_axes()`       | Agrega un nuevo conjunto de ejes a la figura.                                                              |
| `fig.add_subplot()`    | Agrega un nuevo subgráfico (conjunto de ejes) a la figura.                                                 |
| `fig.subplots()`       | Crea una cuadrícula de subgráficos en la figura y devuelve un arreglo de ejes.                            |
| `fig.suptitle()`       | Agrega un título principal (suptitle) a la figura.                                                        |
| `fig.tight_layout()`   | Ajusta automáticamente los espacios entre subgráficos y los márgenes de la figura para un diseño óptimo.  |
| `fig.show()`           | Muestra la figura actual.                                                                                  |
| `fig.savefig()`        | Guarda la figura en un archivo con formato especificado (por ejemplo, PNG, PDF).                          |

```python
fig = plt.figure()   # Creamos una figura (canvas vacío)
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9])   # Añadimos los ejes con coordenadas y tamaño relativo
axes.plot(x, y, 'b')   # Graficamos x vs y usando el color azul ('b')
```

## plt.subplots()

subplots es un método que crea una figura y un conjunto de ejes (subplots) 

| Comando              | Descripción                                                                                     |
|----------------------|-------------------------------------------------------------------------------------------------|
| `fig, ax = plt.subplots()` | Crea una nueva figura y un conjunto de ejes (subplots). `fig` representa la figura y `ax` es un array o matriz de ejes (subplots). |
| `ax.plot(x, y)`         | Crea un gráfico en los ejes especificados con los datos de `x` e `y`.                        |
| `ax.scatter(x, y)`      | Crea un gráfico de dispersión en los ejes con los datos de `x` e `y`.                         |
| `ax.bar(x, height)`     | Crea un gráfico de barras en los ejes con valores de `x` y alturas especificados.             |
| `ax.hist(x, bins)`      | Crea un histograma en los ejes con los datos de `x` y el número de intervalos `bins`.        |
| `ax.pie(x)`             | Crea un gráfico de torta en los ejes con valores proporcionales especificados en `x`.         |

```python
fig, axes = plt.subplots(nrows=1, ncols=2)   # Creamos una figura con 1 fila y 2 columnas de ejes

axes[0].plot(x, y, 'b')   # Graficamos en el primer eje
axes[1].plot(y, x, 'r')   # Graficamos en el segundo eje
```

Armónicos de sin(x)

```python
y = np.zeros(100) # Creamos un array de 100 ceros para el eje y 
x = np.linspace(0, 20, 100) # Creamos un array de 100 valores entre 0 y 20 para el eje x 
n = np.linspace(1,9,9,endpoint=True) # Creamos un array de 9 valores entre 1 y 9 para los armónicos, endpoint=True indica que el último valor es 9.

harmonics = []

for idx, element in enumerate(n): # Iteramos sobre los armónicos 
    y = np.sin(x*element) # Calculamos el seno de cada armónico
    harmonics.append(y) # Agregamos el armónico a la lista de armónicos

harmonics = np.array(harmonics) # Convertimos la lista de armónicos a un array de numpy 
harmonics = np.expand_dims(harmonics, axis=0) # Agregamos una dimensión al array de armónicos 
harmonics = harmonics.reshape((3,3,100)) # Redimensionamos el array de armónicos a una matriz de 3x3x100

nrows = 3 # Número de filas
ncols = 3 # Número de columnas


fig, axes = plt.subplots(nrows, ncols, figsize = (12,6)) # Creamos una figura con 3 filas y 3 columnas de ejes, con tamaño de figura de 12x6 

count = 0 # Contador para los armónicos
 
for idx in range(nrows): # Iteramos sobre las filas
    for jdx in range(ncols): # Iteramos sobre las columnas
        count +=1 # Incrementamos el contador
        axes[idx, jdx].plot(x, harmonics[idx, jdx]) # Graficamos el armónico
        axes[idx, jdx].set_title(f'Harmonic of $sin(%sx)$' % count) # Establecemos el título del armónico

fig.tight_layout()
```

```python
fig, ((axes1, axes2, axes3, axes4), (axes5, axes6, axes7, axes8)) = plt.subplots(2,4)  # Creamos una figura con 2 filas y 4 columnas de ejes 
axes1.plot(x, np.sin(x), 'b')  # Graficamos en el primer eje de la primera fila 
axes2.plot(y, np.cos(x), 'purple') # Graficamos en el segundo eje de la primera fila
axes3.plot(x, np.tan(y), 'orange')
axes3.set_title('Relación X - Y')
axes4.plot(y, np.cos(y)**2, 'black')
axes5.hist(np.sin(x))
axes6.pie(x)
axes7.plot(y, np.sin(y), 'orange')
axes8.boxplot(y,'y')


#mejorar la visualización de las figuras
fig.tight_layout()
```

```python
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p): # función que calcula el potencial del qubit de flujo 
    return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p) # retorna el potencial del qubit de flujo 
```

```python
phi_m = np.linspace(0, 2*np.pi, 100) # generamos 100 valores entre 0 y 2pi para phi_m 
phi_p = np.linspace(0, 2*np.pi, 100) # generamos 100 valores entre 0 y 2pi para phi_p 
X,Y = np.meshgrid(phi_p, phi_m) # generamos una malla de 100x100 con los valores de phi_p y phi_m 
Z = flux_qubit_potential(X, Y).T # calculamos el potencial del qubit de flujo para cada par de valores de phi_p y phi_m
```

```python
fig, ax = plt.subplots()

p = ax.pcolor(X/(2*np.pi), Y/(2*np.pi), Z, cmap=plt.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max()) # cmap es el mapa de colores, vmin y vmax son los valores mínimo y máximo de la escala de colores
# pcolor es una forma de graficar superficies en 2D     
cb = fig.colorbar(p, ax=ax)
```
