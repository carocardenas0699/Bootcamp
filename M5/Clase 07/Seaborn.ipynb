# Seaborn: statistical data visualization

**Seaborn** es una biblioteca construida sobre Matplotlib, heredando todas sus ventajas. Fue desarrollada por Michael Waskom y está diseñada para trabajar de manera óptima con DataFrames de Pandas.

## Ventajas de Seaborn

Seaborn ofrece diversas ventajas, algunas de las cuales son:

- **Rendimiento**: Funciona con gran velocidad.
- **Facilidad de Uso**: Escribir código se vuelve sencillo.
- **Personalización Avanzada**: Permite una alta personalización en gráficas y visualizaciones.
- **Visualización de Relaciones Estadísticas**: Permite visualizar relaciones estadísticas entre variables.
- **Sintaxis más sencilla:** La sintaxis de Seaborn es más concisa y fácil de entender que la de matplotlib.
- **Gráficos más atractivos:** Seaborn proporciona una serie de temas y estilos predeterminados que hacen que los gráficos sean más atractivos.

**Desventajas de Seaborn**

Seaborn también tiene algunas desventajas, entre las que se incluyen:

- **Requiere matplotlib:** Seaborn se basa en matplotlib, por lo que es necesario instalar matplotlib para poder usar Seaborn.
- **Puede ser menos flexible que matplotlib:** Seaborn proporciona una serie de funciones predefinidas, pero puede ser menos flexible que matplotlib si se necesita crear gráficos personalizados.



Seaborn es una buena opción para crear gráficos estadísticos atractivos y fáciles de entender. Es una buena opción para los siguientes casos:

- **Visualización de datos para principiantes:** Seaborn es una buena opción para los principiantes en visualización de datos, ya que proporciona una interfaz sencilla y fácil de usar.
- **Visualización de datos para análisis exploratorio:** Seaborn es una buena opción para el análisis exploratorio de datos, ya que proporciona una serie de funciones para explorar los datos de forma visual.
- **Visualización de datos para presentaciones:** Seaborn es una buena opción para la creación de gráficos para presentaciones, ya que proporciona una serie de temas y estilos predeterminados que hacen que los gráficos sean más atractivos.

## Tipos de Gráficas en Seaborn

Seaborn proporciona una variedad de funciones para abordar diferentes tipos de visualización:

- **Gráficos Relacionales**: `relplot` (scatterplot, lineplot).
- **Distribución de Datos**: `displot` (histplot, kdeplot, ecdfplot, rugplot).
- **Gráficos de Variables Categóricas**: `catplot` (stripplot, swamplot, boxplot, violinplot, pointplot, barplot).

### Crear distribuciones

| Tipo de Gráfico              | Función de Seaborn       | Descripción                                                |
|------------------------------|--------------------------|------------------------------------------------------------|
| Gráfico de Histograma        | `sns.histplot`           | Muestra la distribución de una variable numérica.          |
| Gráfico de KDE               | `sns.kdeplot`            | Estima y visualiza la función de densidad de kernel.       |
| Gráfico de ECDF              | `sns.ecdfplot`           | Muestra la función de distribución acumulativa empírica.   |

````python
# Crea un histograma utilizando Seaborn y muestra distribuciones de propinas

sns.histplot(data=tips, x='tip',bins=15,cumulative=False,hue='sex',stat='count',multiple='dodge') 
# El parámetro "stat" me permite definir el tipo de estadística a mostrar: "count", "frequency", "density", "probability"
plt.show()
````

````python
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count') # Crea un gráfico de densidad de distribución acumulada utilizando Seaborn con la función ecdfplot
plt.show()
````

````python
sns.displot(data=tips, x='tip', hue='smoker', kind='kde', legend=True, palette='dark', alpha=.9) # Crea un gráfico de distribución utilizando Seaborn con la función displot
plt.show()
````

````python
# Crea un histograma utilizando Seaborn y muestra distribuciones de propinas por sexo
sns.displot(data = tips, x='tip', hue='sex', kind='hist', multiple='stack') # stack = apila ambos histogramas, bins viene por defecto en 10
````

## Al igual que con displot, se puede usar sns.catplot()

sns.catplot() define el mejor gráfico de acuerdo a nuestros datos.

### Parámetros de sns.catplot

| Parámetro       | Descripción                                                                   |
|-----------------|-------------------------------------------------------------------------------|
| `data`          | Conjunto de datos (DataFrame) para la visualización.                         |
| `x`             | Variable en el eje x para agrupar y categorizar.                              |
| `y`             | Variable en el eje y para representar valores.                                |
| `hue`           | Variable categórica para distinguir elementos por color.                      |
| `col`           | Variable categórica para dividir el gráfico en columnas.                       |
| `row`           | Variable categórica para dividir el gráfico en filas.                          |
| `kind`          | Tipo de gráfico categórico a crear ('strip', 'swarm', 'box', 'violin', etc.). |
| `order`         | Orden de las categorías en el eje x.                                          |
| `col_order`     | Orden de las categorías en la división por columnas.                          |
| `row_order`     | Orden de las categorías en la división por filas.                             |
| `orient`        | Orientación del gráfico ('v' para vertical, 'h' para horizontal).             |
| `height`        | Altura de cada aspecto del gráfico.                                           |
| `aspect`        | Relación de aspecto de cada aspecto del gráfico.                               |
| `palette`       | Paleta de colores para elementos categóricos.                                 |
| `dodge`         | Si los elementos deben separarse en la misma categoría.                        |
| `share{x,y}`    | Compartir ejes entre aspectos (True o False).                                  |
| `legend`        | Mostrar la leyenda (True o False).                                            |
| `legend_out`    | Posición de la leyenda ('True', 'brief' o ubicación).                          |

````python
# Crea un gráfico categórico utilizando Seaborn y muestra un conjunto de diagramas de caja
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,kind='box',col='time') 
                                        # kind='box' me permite definir el tipo de gráfico categórico a utilizar en este caso diagramas de caja
                                        # col='time' me permite dividir el gráfico por la variable time

plt.show()
````

````python
sns.catplot(data = tips, x="day", y ='total_bill', hue="smoker", col="time", dodge=True)
                    #col="time" me permite dividir el gráfico por la variable time

plt.show()
````

````python
plt.figure(figsize=(8,8))
markers = {"Lunch":"D", "Dinner":"s"} #Definir marcadores en un diccionario
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers) # style me permite definir la variable que define el estilo de los puntos
                                   # size me permite definir la variable que define el tamaño de los puntos
                                      # markers me permite definir el diccionario de marcadores
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
````

## sns.heatmap()

Es una representación visual de una matriz de datos numéricos donde los valores individuales se mapean a colores. Este tipo de gráfico es útil para visualizar patrones y relaciones en datos tabulares.

````python
sns.heatmap(tips.corr())
plt.show()
````

### Se le puede dar más contexto al gráfico

| Parámetro           | Descripción                                                                                      |
|---------------------|--------------------------------------------------------------------------------------------------|
| `data`              | Conjunto de datos (DataFrame) para la visualización.                                            |
| `annot`             | Si es `True`, muestra los valores dentro de las celdas.                                         |
| `fmt`               | Formato de cadena para los valores en `annot`.                                                  |
| `cmap`              | Mapa de colores para representar los valores.                                                   |
| `center`            | Valor alrededor del cual se centrarán los colores del mapa.                                     |
| `cbar`              | Si es `True`, muestra la barra de color a un lado del heatmap.                                  |
| `cbar_kws`          | Argumentos pasados a la función `cbar_ax.colorbar()` si `cbar` es `True`.                        |
| `linewidths`        | Ancho de las líneas entre las celdas.                                                            |
| `linecolor`         | Color de las líneas entre las celdas.                                                            |
| `square`            | Si es `True`, hace que las celdas sean cuadradas.                                                |
| `xticklabels`, `yticklabels` | Si son `True`, muestra etiquetas en los ejes x e y.                                        |
| `cbar_ax`           | Eje en el que se dibuja la barra de color.                                                      |
| `mask`              | Máscara booleana para ocultar valores en el heatmap.                                            |
| `ax`                | Eje en el que se dibujará el heatmap.                                                           |
| `vmin`, `vmax`      | Valores mínimo y máximo para el mapeo de colores.                                               |
| `robust`            | Si es `True`, ajusta el rango de mapeo de colores para evitar valores extremos.                  |
| `square`            | Si es `True`, hace que las celdas sean cuadradas.                                                |

````python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm',linewidths=5,linecolor='black', # annot me permite mostrar los valores de correlación en el mapa de calor,
                                                                                # cmap me permite definir la paleta de colores a utilizar en el mapa de calor,
                                                                                # linewidths me permite definir el ancho de las líneas del mapa de calor,
                                                                                # linecolor me permite definir el color de las líneas del mapa de calor,
                                                                                
            vmin=0.5,vmax=1,cbar=True)# cbar me permite mostrar la barra de color en el mapa de calor
plt.show()
````
